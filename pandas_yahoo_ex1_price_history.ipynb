```python
import pandas_datareader as pdr
import pandas as pd
import datetime
import plotly.express as px

#*****************************************************************************
#
# Relies on yahoo finance api which is unsupported & sometimes doesn't work.
#
#*****************************************************************************

start = datetime.datetime (2005, 5, 1)
end = datetime.datetime (2020, 6, 1)

df = pdr.DataReader('AMZN', 'yahoo', start, end)

# display(df)

fig = px.line(df, y='Close',
              title='Amazon Price Chart 2005-2022',
              labels={'PRICE':'US Dollars'})
fig.show()
```
